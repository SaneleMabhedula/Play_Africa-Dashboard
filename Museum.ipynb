```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
import plotly.express as px
import streamlit as st
plt.style.use('default')  # Use default (light) style
plt.style.use('seaborn-v0_8-whitegrid')
plt.savefig("my_plot.png", facecolor="white")


```

```python
df8 = pd.read_csv(r"C:\Users\CDP USER\Desktop\Python docs\program_impact_measurement.csv", delimiter = ";")
df8
```

```python
# Create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df8, x="Program Name", y="No. of Participants")
# Rotate program names for clarity
plt.xticks(rotation=45);
```

```python




st.title("Children's Museum Program Analysis")

# Show table
st.subheader("Raw Data")
st.dataframe(df8)

# Show bar chart of program ratings
st.subheader("Average Ratings per Program")
avg_ratings = df8.groupby("Program Name")["Program Rating (1-5)"].mean()
st.bar_chart(avg_ratings)

# Show boxplot
st.subheader("Rating Distribution by Program")
fig, ax = plt.subplots()
sns.boxplot(x="Program Name", y="Program Rating (1-5)", data=df8, ax=ax)
plt.xticks(rotation=45)
st.pyplot(fig)
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x="Program Name", y="Program Rating (1-5)", data=df8)
plt.xticks(rotation=45)
plt.title("Average Program Ratings per Program")
plt.ylabel("Rating (1–5)")
plt.tight_layout()
plt.show()
```
